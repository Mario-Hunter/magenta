```

# From https://github.com/pytorch/examples/blob/master/fast_neural_style/neural_style/vgg.py
class Vgg16(tf.keras.Model):
    def __init__(self, device='cpu'):
        super(Vgg16, self).__init__()
        vgg_pretrained_features = VGG16(weights='imagenet', include_top=False).layers
        self.slice1 = Sequential()
        self.slice2 = Sequential()
        self.slice3 = Sequential()
        self.slice4 = Sequential()
        for x in range(4):
            self.slice1.add(vgg_pretrained_features[x])
        for x in range(4, 7):
            self.slice2.add(vgg_pretrained_features[x])
        for x in range(7, 11):
            self.slice3.add(vgg_pretrained_features[x])
        for x in range(11, 15):
            self.slice4.add(vgg_pretrained_features[x])
        
        for param in self.layers:
            param.requires_grad = False

    def call(self, X,  training=False):
        h = self.slice1(X)
        h_relu1_2 = h
        h = self.slice2(h)
        h_relu2_2 = h
        h = self.slice3(h)
        h_relu3_3 = h
        h = self.slice4(h)
        h_relu4_3 = h
        # vgg_outputs = namedtuple("VggOutputs", ['relu1_2', 'relu2_2', 'relu3_3', 'relu4_3'])
        # out = vgg_outputs(h_relu1_2, h_relu2_2, h_relu3_3, h_relu4_3)
        return (h_relu1_2, h_relu2_2, h_relu3_3, h_relu4_3)
```

## Warp

```
def warp(x, flo, Height=None, Width=None):
    """
    warp an image/tensor (im2) back to im1, according to the optical flow
    x: [B, C, H, W] (im2)
    flo: [B, 2, H, W] flow
    """
    if (len(x.shape) == 3):
      x = tf.expand_dims(x, 0)
      
    B, H, W, C =  x.shape 
    
    if Height is not None:
      H = Height
    if Width is not None:
      W = Width

    # mesh grid
    xx = tf.tile(tf.expand_dims(tf.range(0, W), axis=0),(H, 1))
    yy = tf.tile(tf.transpose(tf.expand_dims(tf.range(0, H), axis = 0), (1,0) ), (1, W))
    xx = tf.tile( tf.expand_dims(tf.expand_dims(xx, axis = 0 ), axis = 0), (B, 1, 1, 1) )
    yy = tf.tile( tf.expand_dims(tf.expand_dims(yy, axis = 0 ), axis = 0), (B, 1, 1, 1) )
    grid = tf.transpose(tf.concat((xx, yy), 1), (0,2,3,1)) # channel last
    
    # if x.is_cuda:
    #     grid = grid.cuda()
    vgrid = tf.cast(grid, tf.float32) + flo

    print(vgrid.shape)
    # scale grid to [-1,1]
    ch1 = 2.0 * vgrid[:, :, :, 0] / max(W - 1, 1) - 1.0
    ch2 = 2.0 * vgrid[:, :, :, 1] / max(H - 1, 1) - 1.0

    vgrid = tf.stack([ch1,ch2], axis=3)
    
    output = dense_image_wrap(x, vgrid)
    mask = tf.ones((B,H,W,C))
    mask = dense_image_wrap(mask, vgrid)

    # if W==128:
    # np.save('mask.npy', mask.cpu().data.numpy())
    # np.save('warp.npy', output.cpu().data.numpy())

    mask = tf.greater(mask, tf.constant(0.9999))
    mask = tf.cast(mask, output.dtype)

    return output * mask

def interpolate_bilinear(grid, query_points, indexing="ij", name=None):
    """Similar to Matlab's interp2 function.
    Finds values for query points on a grid using bilinear interpolation.
    Args:
      grid: a 4-D float `Tensor` of shape `[batch, height, width, channels]`.
      query_points: a 3-D float `Tensor` of N points with shape
        `[batch, N, 2]`.
      indexing: whether the query points are specified as row and column (ij),
        or Cartesian coordinates (xy).
      name: a name for the operation (optional).
    Returns:
      values: a 3-D `Tensor` with shape `[batch, N, channels]`
    Raises:
      ValueError: if the indexing mode is invalid, or if the shape of the
        inputs invalid.
    """
    if indexing != "ij" and indexing != "xy":
        raise ValueError("Indexing mode must be \'ij\' or \'xy\'")

    with tf.name_scope(name or "interpolate_bilinear"):
        grid = tf.convert_to_tensor(grid)
        query_points = tf.convert_to_tensor(query_points)

        # grid shape checks
        grid_static_shape = grid.shape
        grid_shape = tf.shape(grid)
        if grid_static_shape.dims is not None:
            if len(grid_static_shape) != 4:
                raise ValueError("Grid must be 4D Tensor")
            if grid_static_shape[1] is not None and grid_static_shape[1] < 2:
                raise ValueError("Grid height must be at least 2.")
            if grid_static_shape[2] is not None and grid_static_shape[2] < 2:
                raise ValueError("Grid width must be at least 2.")
        else:
            # pylint: disable=bad-continuation
            with tf.control_dependencies([
                    tf.debugging.assert_greater_equal(
                        grid_shape[1],
                        2,
                        message="Grid height must be at least 2."),
                    tf.debugging.assert_greater_equal(
                        grid_shape[2],
                        2,
                        message="Grid width must be at least 2."),
                    tf.debugging.assert_less_equal(
                        tf.cast(
                            grid_shape[0] * grid_shape[1] * grid_shape[2],
                            dtype=tf.dtypes.float32),
                        np.iinfo(np.int32).max / 8.0,
                        message="The image size or batch size is sufficiently "
                        "large that the linearized addresses used by "
                        "tf.gather may exceed the int32 limit.")
            ]):
                pass
            # pylint: enable=bad-continuation

        # query_points shape checks
        query_static_shape = query_points.shape
        query_shape = tf.shape(query_points)
        if query_static_shape.dims is not None:
            if len(query_static_shape) != 3:
                raise ValueError("Query points must be 3 dimensional.")
            query_hw = query_static_shape[2]
            if query_hw is not None and query_hw != 2:
                raise ValueError("Query points last dimension must be 2.")
        else:
            # pylint: disable=bad-continuation
            with tf.control_dependencies([
                    tf.debugging.assert_equal(
                        query_shape[2],
                        2,
                        message="Query points last dimension must be 2.")
            ]):
                pass
            # pylint: enable=bad-continuation

        batch_size, height, width, channels = (grid_shape[0], grid_shape[1],
                                               grid_shape[2], grid_shape[3])

        num_queries = query_shape[1]

        query_type = query_points.dtype
        grid_type = grid.dtype

        alphas = []
        floors = []
        ceils = []
        index_order = [0, 1] if indexing == "ij" else [1, 0]
        unstacked_query_points = tf.unstack(query_points, axis=2, num=2)

        for i, dim in enumerate(index_order):
            with tf.name_scope("dim-" + str(dim)):
                queries = unstacked_query_points[dim]

                size_in_indexing_dimension = grid_shape[i + 1]

                # max_floor is size_in_indexing_dimension - 2 so that max_floor + 1
                # is still a valid index into the grid.
                max_floor = tf.cast(size_in_indexing_dimension - 2, query_type)
                min_floor = tf.constant(0.0, dtype=query_type)
                floor = tf.math.minimum(
                    tf.math.maximum(min_floor, tf.math.floor(queries)),
                    max_floor)
                int_floor = tf.cast(floor, tf.dtypes.int32)
                floors.append(int_floor)
                ceil = int_floor + 1
                ceils.append(ceil)

                # alpha has the same type as the grid, as we will directly use alpha
                # when taking linear combinations of pixel values from the image.
                alpha = tf.cast(queries - floor, grid_type)
                min_alpha = tf.constant(0.0, dtype=grid_type)
                max_alpha = tf.constant(1.0, dtype=grid_type)
                alpha = tf.math.minimum(
                    tf.math.maximum(min_alpha, alpha), max_alpha)

                # Expand alpha to [b, n, 1] so we can use broadcasting
                # (since the alpha values don't depend on the channel).
                alpha = tf.expand_dims(alpha, 2)
                alphas.append(alpha)

            flattened_grid = tf.reshape(
                grid, [batch_size * height * width, channels])
            batch_offsets = tf.reshape(
                tf.range(batch_size) * height * width, [batch_size, 1])
        # pylint: enable=bad-continuation

        # This wraps tf.gather. We reshape the image data such that the
        # batch, y, and x coordinates are pulled into the first dimension.
        # Then we gather. Finally, we reshape the output back. It's possible this
        # code would be made simpler by using tf.gather_nd.
        def gather(y_coords, x_coords, name):
            with tf.name_scope("gather-" + name):
                linear_coordinates = (
                    batch_offsets + y_coords * width + x_coords)
                gathered_values = tf.gather(flattened_grid, linear_coordinates)
                return tf.reshape(gathered_values,
                                  [batch_size, num_queries, channels])

        # grab the pixel values in the 4 corners around each query point
        top_left = gather(floors[0], floors[1], "top_left")
        top_right = gather(floors[0], ceils[1], "top_right")
        bottom_left = gather(ceils[0], floors[1], "bottom_left")
        bottom_right = gather(ceils[0], ceils[1], "bottom_right")

        # now, do the actual interpolation
        with tf.name_scope("interpolate"):
            interp_top = alphas[1] * (top_right - top_left) + top_left
            interp_bottom = alphas[1] * (
                bottom_right - bottom_left) + bottom_left
            interp = alphas[0] * (interp_bottom - interp_top) + interp_top

        return interp

def dense_image_wrap(image, flow):
  image = tf.convert_to_tensor(image)
  flow = tf.convert_to_tensor(flow)
  batch_size, height, width, channels = (tf.shape(image)[0],
                                          tf.shape(image)[1],
                                          tf.shape(image)[2],
                                          tf.shape(image)[3])

  # The flow is defined on the image grid. Turn the flow into a list of query
  # points in the grid space.
  grid_x, grid_y = tf.meshgrid(tf.range(width), tf.range(height))
  stacked_grid = tf.cast(tf.stack([grid_y, grid_x], axis=2), flow.dtype)
  batched_grid = tf.expand_dims(stacked_grid, axis=0)
  query_points_on_grid = batched_grid - flow
  query_points_flattened = tf.reshape(query_points_on_grid,
                                      [batch_size, height * width, 2])
  # Compute values at the query points, then reshape the result back to the
  # image grid.
  interpolated = interpolate_bilinear(image, query_points_flattened)
  interpolated = tf.reshape(interpolated,
                            [batch_size, height, width, channels])
  return interpolated
```

```
from magenta.models.image_stylization import vgg as img_vgg
from tensorflow.python.ops import variables as tf_variables
# sess = tf.Session()
# sess.graph._unsafe_unfinalize()

img1_ph, img1_preprocessed = img_placeholder(content_square_crop, "styling/frame1")
img2_ph, img2_preprocessed = img_placeholder(content_square_crop, "styling/frame2")

styled_img1, bottleneck1, feature_map1 = forward(img1_preprocessed, style_img_preprocessed)
styled_img2, bottleneck2, feature_map2 = forward(img2_preprocessed, style_img_preprocessed)

img1_preprocessed_norm, img2_preprocessed_norm = tf_normalize(img1_preprocessed), tf_normalize(img2_preprocessed)
styled_img1_norm, styled_img2_norm = tf_normalize(styled_img1), tf_normalize(styled_img2)

mask_ph = tf.placeholder(tf.float32, shape=[1, 360, 640, 1], name= "styling/mask")
flow_ph = tf.placeholder(tf.float32, shape=[1, 360, 640, 2], name= "styling/flow")



vgg =  Vgg16()

styled_features1 = vgg(styled_img1)
styled_features2 = vgg(styled_img2)
img_features1 = vgg(img1_preprocessed)
img_features2 = vgg(img2_preprocessed)
# print(styled_features1[0].shape)
feature_map_size = (90,160)
styled_features_shape = (1, 180, 320, 64)


feature_flow = tf.image.resize_bilinear( flow_ph , feature_map_size)
feature_mask = tf.expand_dims(tf.image.resize_bilinear(mask_ph,feature_map_size)[:,:,:,0], axis=-1)
warped_fmap = warp(feature_map1, feature_flow, feature_map_size[0], feature_map_size[1])

#Changed by KJ to multiply with feature mask
# # print(L2distancematrix(feature_map2, warped_fmap).size()) #Should be a matrix not number
# # mean replaced sum
f_temporal_loss = tf.math.reduce_sum(tf.broadcast_to(feature_mask, (1, feature_map_size[0], feature_map_size[1], feature_map2.shape[-1])) * (L2distancematrix(feature_map2, warped_fmap)))
f_temporal_loss *= LAMBDA_F
f_temporal_loss *= 1 / (feature_map_size[0] * feature_map_size[1] * feature_map2.shape[3].value)

# # Removed unsqueeze methods in both styled_img1,flow in next line since already 4 dimensional
warped_style = warp(styled_img1, flow_ph, 360, 640)
warped_image = warp(img1_ph, flow_ph, 360, 640)


output_term = styled_img2[0] - warped_style[0]
input_term = img2_ph[0] - warped_image[0]

# Changed the next few lines since dimension is 4 instead of 3 with batch size=1
input_term = 0.2126 * input_term[:, :, 0] + 0.7152 * input_term[:, :, 1] + 0.0722 * input_term[:, :, 2]
input_term = tf.stack([input_term,input_term,input_term], axis =-1)


o_temporal_loss = tf.math.reduce_sum(mask_ph * (L2distancematrix(output_term, input_term)))
o_temporal_loss *= LAMBDA_O
o_temporal_loss *= 1 / (360 * 640)

estab = tf.reduce_mean(mask_ph * (L2distancematrix(styled_img2, warped_style))) 

content_loss = 0
# content_loss += L2distance(styled_features1[1],  img_features1[1])
# content_loss += L2distance(styled_features2[1], img_features2[1])
content_loss += tf.reduce_mean((styled_features1[2] - img_features1[2]) ** 2)
content_loss += tf.reduce_mean((styled_features2[2] - img_features2[2]) ** 2)
content_loss *= ALPHA / (  styled_features_shape[1] * styled_features_shape[2] *styled_features_shape[3])

style_loss = 0

if 'style_GM' not in locals():
  # styled_featuresR = vgg(tf_normalize(style))
  styled_featuresR = vgg(tf.cast(style/255.0, tf.float32))
  style_GM = [tf_gram_matrix(f) for f in styled_featuresR]

for i, weight in enumerate(STYLE_WEIGHTS):
  gram_s = style_GM[i]
  gram_img1 = tf_gram_matrix(styled_features1[i])
  gram_img2 = tf_gram_matrix(styled_features2[i])
  # style_loss += float(weight) * (L2distance(gram_img1, gram_s) + L2distance(gram_img2, gram_s))
  style_loss += float(weight) * tf.reduce_mean((gram_img1 - gram_s) ** 2)
  style_loss += float(weight) * tf.reduce_mean((gram_img2 - gram_s) ** 2)
style_loss *= BETA

reg_loss = GAMMA * \
  (tf.math.reduce_sum(tf.math.abs(styled_img1[:, :, :, :-1] - styled_img1[:, :, :, 1:])) +
    tf.math.reduce_sum(tf.math.abs(styled_img1[:, :, :-1, :] - styled_img1[:, :, 1:, :])))

reg_loss += GAMMA * \
    (tf.math.reduce_sum(tf.math.abs(styled_img2[:, :, :, :-1] - styled_img2[:, :, :, 1:])) +
    tf.math.reduce_sum(tf.math.abs(styled_img2[:, :, :-1, :] - styled_img2[:, :, 1:, :])))

loss = f_temporal_loss + o_temporal_loss + content_loss + style_loss + reg_loss

# Function to restore VGG16 parameters.
init_fn_vgg = slim.assign_from_checkpoint_fn('/content/vgg_16.ckpt',
                                              slim.get_variables())

# Function to restore Inception_v3 parameters.
inception_variables_dict = {
    var.op.name: var
    for var in slim.get_model_variables('InceptionV3')
}
init_fn_inception = slim.assign_from_checkpoint_fn(
    '/content/inception_v3.ckpt', inception_variables_dict)

# Function to restore VGG16 and Inception_v3 parameters.
def init_sub_networks(session):
    # init_fn_vgg(session)
    init_fn_inception(session)


# train_op = slim.learning.create_train_op(
#             loss,
#             tf.train.AdamOptimizer(1e-3),
#             clip_gradient_norm=False,
#             summarize_gradients=False)
optimizer = tf.train.AdamOptimizer(learning_rate)

# grads = optimizer.compute_gradients(loss, tf_variables.trainable_variables())
# grad_updates = optimizer.apply_gradients(grads)
# slim.learning.train(
#   train_op=train_op,
#   init_feed_dict= {img1_ph.name:tf.transpose(img1[0], (1,2,0)).eval(), style_ph.name:style[0], img2_ph.name:tf.transpose(img2[0], (1,2,0)).eval()},
#   init_fn=init_sub_networks,
#   logdir=os.path.expanduser("train_logs/"),
#   number_of_steps=1)
with tf.variable_scope('', reuse=tf.AUTO_REUSE):
  grads = optimizer.compute_gradients(loss, var_list=tf.get_collection(tf.GraphKeys.GLOBAL_VARIABLES, scope='transformer'))
  train_op = optimizer.minimize(loss, var_list=tf.get_collection(tf.GraphKeys.GLOBAL_VARIABLES, scope='transformer'))


```

```
start_i = 1
start_ep = 4

dataloader = DataLoader(MPIDataset("./", offset= start_i), batch_size=1, )
```

```
row_format ="{:>5}{:^12}" + ("{:^24}" * (5)) + "{:>12}"
row_numbers_format ="{:<5}{:^12}" + ("{:^9.3f}/{:^15.3f}" * (4)) + "{:^24.0f}{:>12.4f}"
t_losses, s_losses, c_losses, f_t_losses, o_t_losses, stab_losses = [], [], [], [], [],[]
epochs = 10

sess = tf.Session()
step = tf.train.get_or_create_global_step(graph=sess.graph)
init_sub_networks(sess)
sess.run(tf.global_variables_initializer())
maybe_initialize(checkpoint, initialized, sess)


trained_ckpt = "/content/gdrive/My Drive/College Deliverables /CS-4/Deep Learning/Project/checkpoints/model-"+str(start_ep)+"-"+str(100)+".ckpt"

saver = tf.train.Saver(tf.get_collection(tf.GraphKeys.GLOBAL_VARIABLES, 'transformer'))
saver.restore(sess, trained_ckpt)

print(row_format.format(*["iteration", "Frame", "Style Loss", "Content Loss", "F-Temporal Loss", "O-Temporal Loss", "Total loss", "Stability Error"]))
# saver.save(sess, "train_dir/model.ckpt", global_step=step)
for ep in range(start_ep, epochs):
  for i, (folder,num, img1, img2, mask, flow) in enumerate(dataloader):
    
    flow = -flow
    img1 = img1
    # img1 = (img1 - VGG16_MEAN)/VGG16_STD
    img2 = img2
    # img2 = (img2 - VGG16_MEAN)/VGG16_STD
    identity1 = sess.run(bottleneck1, feed_dict={style_ph: img1[0]})
    identity2 = sess.run(bottleneck2, feed_dict={style_ph: img2[0]})
    style_params = sess.run(bottleneck1, feed_dict={style_ph: style[0]})
    b1_value = identity1*0.7 + 0.3* style_params
    b2_value =  identity2*0.7 + 0.3* style_params
    s1 = sess.run(styled_img1, feed_dict={img1_ph:img1[0], style_ph:style[0], bottleneck1: b1_value})
    s2 = sess.run(styled_img2, feed_dict={img2_ph:img2[0], style_ph:style[0],  bottleneck2: b2_value })

    _,  t_loss, s_loss, c_loss, f_t_loss, o_t_loss, stab_loss = sess.run([train_op, loss, style_loss, content_loss, f_temporal_loss, o_temporal_loss, estab ], 
                                                            feed_dict={img1_ph:img1[0], style_ph:style[0], img2_ph:img2[0], mask_ph:mask, flow_ph:flow})

    print(row_numbers_format.format(*[i+start_i-1, folder[0]+"-"+num[0], s_loss/BETA, s_loss, c_loss/ALPHA, c_loss, f_t_loss/LAMBDA_F,f_t_loss, o_t_loss/LAMBDA_O, o_t_loss, t_loss, stab_loss]))
    if i % 100 == 0:
      t_losses.append(t_loss)
      s_losses.append(s_loss)
      c_losses.append(c_loss)
      f_t_losses.append(f_t_loss)
      o_t_losses.append(o_t_loss)
      stab_losses.append(stab_loss)
    
      saver.save(sess, "/content/gdrive/My Drive/College Deliverables /CS-4/Deep Learning/Project/checkpoints/model-"+str(ep)+"-"+str(i+start_i-1)+".ckpt")
saver.save(sess, "train_dir/model.ckpt", global_step=step,)

sess.close()
      # saver.save(sess, "train_dir/model.ckpt")
```
